# YOLOv9

## updated dataset

```python
import fiftyone as fo

dataset = fo.load_dataset('new_test_data_yolo9e_conf03_best_model_updated_dataset')
# Get the 10 most common classes in the dataset
counts = dataset.count_values("full_GT.detections.label")
classes = sorted(counts, key=counts.get, reverse=True)[:3]

# Evaluate the objects in the `predictions` field with respect to the
# objects in the `ground_truth` field
results = dataset.evaluate_detections(
    pred_field="preds_yolo9e_conf03_iou03_best_updated_dataset",
    gt_field="full_GT",
    eval_key="full_GT_vs_yolo9x",
    method="coco",
    compute_mAP=True,
    iou=0.25,
    classwise=False
)

# Print a classification report for the top-10 classes
results.print_report(classes=classes)

# Print some statistics about the total TP/FP/FN counts
print("TP: %d" % dataset.sum("full_GT_vs_yolo9x_tp"))
print("FP: %d" % dataset.sum("full_GT_vs_yolo9x_fp"))
print("FN: %d" % dataset.sum("full_GT_vs_yolo9x_fn"))
```

```python
plot = results.plot_confusion_matrix(backend="matplotlib")
plot.show()
```
